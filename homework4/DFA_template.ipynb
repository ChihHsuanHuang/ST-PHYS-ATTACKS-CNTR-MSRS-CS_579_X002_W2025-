```python
import numpy as np


ctext_path = "aes-dfa/ctext.bin"
ftext_path = "aes-dfa/ftext.bin"


ctext = np.fromfile(ctext_path, dtype=np.uint8).reshape(1000, 16)
ftext = np.fromfile(ftext_path, dtype=np.uint8).reshape(1000, 16)


def classify_fault_pairs():
    column_1 = []  
    column_2 = []  
    column_3 = []  
    column_4 = []  

    
    def is_fault_in_column(xor_result, column_index):
       
        column_map = {
            1: [0, 7, 10, 13],
            2: [1, 4, 11, 14],
            3: [2, 5, 8, 15],
            4: [3, 6, 9, 12]
        }

        positions = column_map.get(column_index, [])
        return all(xor_result[pos] != 0 for pos in positions)

   
    for idx in range(1000):
        if np.any(ctext[idx] != ftext[idx]): 
            xor_result = np.bitwise_xor(ctext[idx], ftext[idx]) 

            
            if is_fault_in_column(xor_result, 1):
                column_1.append([ctext[idx], ftext[idx]])
            if is_fault_in_column(xor_result, 2):
                column_2.append([ctext[idx], ftext[idx]])
            if is_fault_in_column(xor_result, 3):
                column_3.append([ctext[idx], ftext[idx]])
            if is_fault_in_column(xor_result, 4):
                column_4.append([ctext[idx], ftext[idx]])

    return column_1, column_2, column_3, column_4

#
def get_faulty_pairs_by_column(pair_count=2):
    col_1, col_2, col_3, col_4 = classify_fault_pairs()

    
    column_1_pairs = col_1[:pair_count]
    column_2_pairs = col_2[:pair_count]
    column_3_pairs = col_3[:pair_count]
    column_4_pairs = col_4[:pair_count]

    return {'column_1': column_1_pairs, 'column_2': column_2_pairs, 'column_3': column_3_pairs, 'column_4': column_4_pairs}


def convert_to_hex(byte_array):
    return ''.join(format(byte, '02x') for byte in byte_array)


def display_pairs_in_hex():
    faulty_pairs = get_faulty_pairs_by_column()

    hex_faulty_pairs = {}
    for col, pairs in faulty_pairs.items():
        hex_faulty_pairs[col] = []
        for c, f in pairs:
            c_hex = convert_to_hex(c)
            f_hex = convert_to_hex(f)
            hex_faulty_pairs[col].append((c_hex, f_hex))

    return hex_faulty_pairs


hex_pairs = display_pairs_in_hex()

for col, pairs in hex_pairs.items():
    print(f"Column {col}:")
    for idx, (c, f) in enumerate(pairs):
        print(f"  Pair {idx + 1}: Ciphertext: {c}, Faulty Ciphertext: {f}")
```

```python
# Galois multiplication by 2 (for MixCols)
def galois_mult_2(a):
    temp = (a << 1) & 0xff

    if (a & 0x80):
        temp ^= 0x1b

    return temp

# Galois multiplication by 3 (for MixCols)
def galois_mult_3(a):
    return galois_mult_2(a) ^ a
```

# Simple Attack Algorithm (finds 4 bytes):

## Variable Definitions
Let $c, c'$ be the first (ciphertext, faultytext) pair and $c, c^{*\prime}$ be the second pair.
Note: $c_0$ refers to the $0$ index byte of the ciphertext

## Preliminary Filtering
1. For each $K_0 \in 0 \text{ to } 255$ and $K_{13} \in 0 \text{ to } 255$, and $x \in 1 \text{ to } 255$
    - Check if $ISBOX(K_0 \oplus c_0) \oplus ISBOX(K_0 \oplus c_0') \stackrel{?}{=} D[x, 0]$
    - Check if $ISBOX(K_{13} \oplus c_{13}) \oplus ISBOX(K_{13} \oplus c_{13}') \stackrel{?}{=} D[x, 1]$

    - If both are true, add candidates $\{K_0, K_{13}\}$ to group $K_{c2}$
2. For each $K_{10} \in 0 \text{ to } 255$, and $\{K_{0}, K_{13}\} \in K_{c2}$, and $x \in 1 \text{ to } 255$
    - Check if $ISBOX(K_0 \oplus c_0) \oplus ISBOX(K_0 \oplus c_0') \stackrel{?}{=} D[x, 0]$
    - Check if $ISBOX(K_{13} \oplus c_{13}) \oplus ISBOX(K_{13} \oplus c_{13}') \stackrel{?}{=} D[x, 1]$
    - Check if $ISBOX(K_{10} \oplus c_{10}) \oplus ISBOX(K_{10} \oplus c_{10}') \stackrel{?}{=} D[x, 2]$

    - If so, add candidates $\{K_0, K_{13}, K_{10}\}$ to group $K_{c3}$
3. For each $K_{7} \in 0 \text{ to } 255$, and $\{K_{0}, K_{13}, K_{10}\} \in K_{c3}$, and $x \in 1 \text{ to } 255$
    - Check if $ISBOX(K_0 \oplus c_0) \oplus ISBOX(K_0 \oplus c_0') \stackrel{?}{=} D[x, 0]$
    - Check if $ISBOX(K_{13} \oplus c_{13}) \oplus ISBOX(K_{13} \oplus c_{13}') \stackrel{?}{=} D[x, 1]$
    - Check if $ISBOX(K_{10} \oplus c_{10}) \oplus ISBOX(K_{10} \oplus c_{10}') \stackrel{?}{=} D[x, 2]$
    - Check if $ISBOX(K_{7} \oplus c_{7}) \oplus ISBOX(K_{7} \oplus c_{7}') \stackrel{?}{=} D[x, 3]$

    - If so, add candidates $\{K_0, K_{13}, K_{10}, K_{7}\}$ to group $K_{c4}$
    
## Finding Final Candidate
1. For each $\{K_{0}, K_{13}, K_{10}, K_7\} \in K_{c4}$, and $x \in 1 \text{ to } 255$
  - $ISBOX(K_0 \oplus c^{*}_0) \oplus ISBOX(K_0 \oplus c^{*\prime}_0) \stackrel{?}{=} D[x, 0]$
  - $ISBOX(K_{13} \oplus c^{*}_{13}) \oplus ISBOX(K_{13} \oplus c^{*\prime}_{13}) \stackrel{?}{=} D[x, 1]$
  - $ISBOX(K_{10} \oplus c^{*}_{10}) \oplus ISBOX(K_{10} \oplus c^{*\prime}_{10}) \stackrel{?}{=} D[x, 2]$
  - $ISBOX(K_{7} \oplus c^{*}_{7}) \oplus ISBOX(K_{7} \oplus c^{*\prime}_{7}) \stackrel{?}{=} D[x, 3]$
  - If all are true $\{K_0, K_{13}, K_{10}, K_{7}\}$, are likely correct keybytes

```python
def precompute_d_table():
    D = []  
    mixcol = MIXCOLS[:, 0]  
    for x in range(1, 256): 
        D_element = []
        for j in range(4): 
            out = None
            if mixcol[j] == 1:
                out = x  
            elif mixcol[j] == 2:
                out = galois_mult_2(x)  
            elif mixcol[j] == 3:
                out = galois_mult_3(x) 
            D_element.append(out)
        D.append(D_element)
    
    return np.array(D)  


D_table = precompute_d_table()

def extract_bytes_from_state(state, faulty_state, indices):
    state_bytes = [state[np.unravel_index(i, shape=(4, 4), order='F')] for i in indices]
    faulty_bytes = [faulty_state[np.unravel_index(i, shape=(4, 4), order='F')] for i in indices]
    
    if any(byte < 0 or byte > 255 for byte in state_bytes + faulty_bytes):
        raise ValueError(f"Extracted bytes are out of valid range: {state_bytes + faulty_bytes}")
    
    return state_bytes, faulty_bytes

def build_lookup_sets(D_table):
    set_2_bytes = {tuple(D_table[i, :2]) for i in range(len(D_table))}
    set_3_bytes = {tuple(D_table[i, :3]) for i in range(len(D_table))}
    set_4_bytes = {tuple(D_table[i, :4]) for i in range(len(D_table))}
    return set_2_bytes, set_3_bytes, set_4_bytes

lookup_set_2, lookup_set_3, lookup_set_4 = build_lookup_sets(D_table)


def simple_dfa_attack_v2(pairs, indices=[0, 7, 10, 13]):
    first_ctxt, first_ftxt = pairs[0]
    second_ctxt, second_ftxt = pairs[1]

    state_bytes, faulty_bytes = extract_bytes_from_state(first_ctxt, first_ftxt, indices)
    state_bytes_x, faulty_bytes_x = extract_bytes_from_state(second_ctxt, second_ftxt, indices)

    candidate_2 = []
    for byte_0 in range(256):
        for byte_13 in range(256):
            try:
                candidate_1 = (ISBOX[byte_0 ^ state_bytes[0]] ^ ISBOX[byte_0 ^ faulty_bytes[0]], 
                               ISBOX[byte_13 ^ state_bytes[3]] ^ ISBOX[byte_13 ^ faulty_bytes[3]])
                if tuple(candidate_1) in lookup_set_2:
                    candidate_2.append((byte_0, byte_13))
            except IndexError:
                continue

    candidate_3 = []
    for byte_10 in range(256):
        for byte_0, byte_13 in candidate_2:
            try:
                candidate_2_values = (ISBOX[byte_0 ^ state_bytes[0]] ^ ISBOX[byte_0 ^ faulty_bytes[0]], 
                                      ISBOX[byte_13 ^ state_bytes[3]] ^ ISBOX[byte_13 ^ faulty_bytes[3]], 
                                      ISBOX[byte_10 ^ state_bytes[2]] ^ ISBOX[byte_10 ^ faulty_bytes[2]])
                if tuple(candidate_2_values) in lookup_set_3:
                    candidate_3.append((byte_0, byte_13, byte_10))
            except IndexError:
                continue

    candidate_4 = []
    for byte_7 in range(256):
        for byte_0, byte_13, byte_10 in candidate_3:
            try:
                candidate_3_values = (ISBOX[byte_0 ^ state_bytes[0]] ^ ISBOX[byte_0 ^ faulty_bytes[0]], 
                                      ISBOX[byte_13 ^ state_bytes[3]] ^ ISBOX[byte_13 ^ faulty_bytes[3]], 
                                      ISBOX[byte_10 ^ state_bytes[2]] ^ ISBOX[byte_10 ^ faulty_bytes[2]], 
                                      ISBOX[byte_7 ^ state_bytes[1]] ^ ISBOX[byte_7 ^ faulty_bytes[1]])
                if tuple(candidate_3_values) in lookup_set_4:
                    candidate_4.append((byte_0, byte_13, byte_10, byte_7))
            except IndexError:
                continue

    candidate_5 = []
    for byte_0, byte_13, byte_10, byte_7 in candidate_4:
        try:
            candidate_4_values = (ISBOX[byte_0 ^ state_bytes_x[0]] ^ ISBOX[byte_0 ^ faulty_bytes_x[0]], 
                                  ISBOX[byte_13 ^ state_bytes_x[3]] ^ ISBOX[byte_13 ^ faulty_bytes_x[3]], 
                                  ISBOX[byte_10 ^ state_bytes_x[2]] ^ ISBOX[byte_10 ^ faulty_bytes_x[2]], 
                                  ISBOX[byte_7 ^ state_bytes_x[1]] ^ ISBOX[byte_7 ^ faulty_bytes_x[1]])
            if tuple(candidate_4_values) in lookup_set_4:
                candidate_5.append((byte_0, byte_13, byte_10, byte_7))
        except IndexError:
            continue

    return candidate_5  


result = simple_dfa_attack_v2([first_pair_simple, second_pair_simple], [0, 7, 10, 13])
print('The key bytes for the simple attack are:', result)
```

```python

def get_bytes_at_positions(ciphertext, faultytext, indices):
    return [
        (ciphertext[i], faultytext[i])  
        for i in indices
    ]


def extract_ciphertext_faultytext_pairs(col_index=None):
    columns = [[] for _ in range(4)]
    counter = 0

    for i in range(1000):
        if not np.all(ctext[i] == ftext[i]):
            xor = np.bitwise_xor(ctext[i], ftext[i])
            if all(xor[j] != 0 for j in [0, 7, 10, 13]):
                columns[0].append([ctext[i], ftext[i]])
                counter += 1
            if all(xor[j] != 0 for j in [1, 4, 11, 14]):
                columns[1].append([ctext[i], ftext[i]])
                counter += 1
            if all(xor[j] != 0 for j in [2, 5, 8, 15]):
                columns[2].append([ctext[i], ftext[i]])
                counter += 1
            if all(xor[j] != 0 for j in [3, 6, 9, 12]):
                columns[3].append([ctext[i], ftext[i]])
                counter += 1

    return columns[col_index - 1] if col_index else columns


def extract_pairs_for_column(pairs_count=2, col_index=None):
    return extract_ciphertext_faultytext_pairs(col_index=col_index)[:pairs_count]


def print_pairs_in_hex_for_column(column_no):
    arr_pairs = extract_pairs_for_column(col_index=column_no)
    return [[[hex(x) for x in ctext], [hex(x) for x in ftext]] for ctext, ftext in arr_pairs]


def get_all_pairs_for_all_columns(pairs_count=2):
    return {f'col_{i+1}': extract_pairs_for_column(pairs_count, i+1) for i in range(4)}

def full_attack_phase(pairs, indices):
    first_ctxt, first_ftxt = pairs[0]
    second_ctxt, second_ftxt = pairs[1]

    bytes_first = get_bytes_at_positions(first_ctxt, first_ftxt, indices)
    bytes_second = get_bytes_at_positions(second_ctxt, second_ftxt, indices)

    candidate_keys = []
    for k0 in range(256):
        for k1 in range(256):
            str_1 = str(ISBOX[k0 ^ bytes_first[0][0]] ^ ISBOX[k0 ^ bytes_first[0][1]]) + str(ISBOX[k1 ^ bytes_first[3][0]] ^ ISBOX[k1 ^ bytes_first[3][1]])
            if str_1 in D2digitsSet :
                candidate_keys.append((k0, k1))

    candidate_keys_phase_2 = []
    for k2 in range(256):
        for k0, k1 in candidate_keys:
            str_2 = str(ISBOX[k0 ^ bytes_first[0][0]] ^ ISBOX[k0 ^ bytes_first[0][1]]) + str(ISBOX[k1 ^ bytes_first[3][0]] ^ ISBOX[k1 ^ bytes_first[3][1]]) + \
                    str(ISBOX[k2 ^ bytes_first[2][0]] ^ ISBOX[k2 ^ bytes_first[2][1]])
            if str_2 in D3digitsSet :
                candidate_keys_phase_2.append((k0, k1, k2))

    candidate_keys_phase_3 = []
    for k3 in range(256):
        for k0, k1, k2 in candidate_keys_phase_2:
            str_3 = str(ISBOX[k0 ^ bytes_first[0][0]] ^ ISBOX[k0 ^ bytes_first[0][1]]) + str(ISBOX[k1 ^ bytes_first[3][0]] ^ ISBOX[k1 ^ bytes_first[3][1]]) + \
                    str(ISBOX[k2 ^ bytes_first[2][0]] ^ ISBOX[k2 ^ bytes_first[2][1]]) + str(ISBOX[k3 ^ bytes_first[1][0]] ^ ISBOX[k3 ^ bytes_first[1][1]])
            if str_3 in D4digitsSet:
                candidate_keys_phase_3.append((k0, k1, k2, k3))

    candidate_keys_phase_4 = []
    for k0, k1, k2, k3 in candidate_keys_phase_3:
        str_4 = str(ISBOX[k0 ^ bytes_second[0][0]] ^ ISBOX[k0 ^ bytes_second[0][1]]) + str(ISBOX[k1 ^ bytes_second[3][0]] ^ ISBOX[k1 ^ bytes_second[3][1]]) + \
                str(ISBOX[k2 ^ bytes_second[2][0]] ^ ISBOX[k2 ^ bytes_second[2][1]]) + str(ISBOX[k3 ^ bytes_second[1][0]] ^ ISBOX[k3 ^ bytes_second[1][1]])
        if str_4 in D4digitsSet:
            candidate_keys_phase_4.append((k0, k1, k2, k3))

    
    if indices == [0, 7, 10, 13]:  
        return [(k0, k1, k2, k3) for k0, k1, k2, k3 in candidate_keys_phase_4]
    elif indices == [1, 4, 11, 14]:  
        return [(k3, k0, k1, k2) for k0, k1, k2, k3 in candidate_keys_phase_4]
    elif indices == [2, 5, 8, 15]:  
        return [(k2, k3, k0, k1) for k0, k1, k2, k3 in candidate_keys_phase_4]
    elif indices == [3, 6, 9, 12]:  
        return [(k1, k2, k3, k0) for k0, k1, k2, k3 in candidate_keys_phase_4]


col_pairs = get_all_pairs_for_all_columns()
result_phase_1 = full_attack_phase(col_pairs['col_1'], [0, 7, 10, 13])
result_phase_2 = full_attack_phase(col_pairs['col_2'], [1, 4, 11, 14])
result_phase_3 = full_attack_phase(col_pairs['col_3'], [2, 5, 8, 15])
result_phase_4 = full_attack_phase(col_pairs['col_4'], [3, 6, 9, 12])

print('First set of key bytes for Phase 1:', result_phase_1)
print('Second set of key bytes for Phase 2:', result_phase_2)
print('Third set of key bytes for Phase 3:', result_phase_3)
print('Fourth set of key bytes for Phase 4:', result_phase_4)
```
